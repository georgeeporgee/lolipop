```python
#монетка, погода, акции, текст в речь, курс валют, переводчик, secret function 
#погода
from bs4 import BeautifulSoup
import requests 
import time
from fake_useragent import UserAgent

url1='https://www.gismeteo.ru/weather-moscow-4368/10-days/'#ходим по страницам сайта (25 страниц по 20 аниме на одной = 500 строк в df)
response = requests.get(url1, headers={'User-Agent': UserAgent().chrome})#используем фейк агент,тк без него код 403
tree =  BeautifulSoup(response.content, 'html.parser')#создаём дерево
source = tree.find_all('div',{'class':'maxt'})
mint=tree.find_all('div',{'class':'mint'})[0].find_all('span')[0].text
maxt=source[0].find_all('span')[0].text
source1 = tree.find_all('div',{'data-item':'0'})
weather=source1[1].span.get('data-text')
source2 = tree.find_all('span',{'class':'unit unit_wind_m_s'})
wind=source2[0].text.strip()
print('Температура сегодня в Москве:от {} до {} . Погода: {}. Средняя скорость ветра:{} м/с'.format(mint,maxt,weather,wind))
```

```python
#акции
import pandas as pd
urrl='https://www.finam.ru/quotes/stocks/russia/?pageNumber=1'
rrr = requests.get(urrl, headers={'User-Agent': UserAgent().chrome})#используем фейк агент,тк без него код 403
frr=pd.read_html(rrr.content)
er=frr[0]
er=er.drop('Unnamed: 0', axis=1)
for p in range(2,16):
    urrl='https://www.finam.ru/quotes/stocks/russia/?pageNumber={}'.format(p)
    rrr = requests.get(urrl, headers={'User-Agent': UserAgent().chrome})#используем фейк агент,тк без него код 403
    frr=pd.read_html(rrr.content)
    rer=frr[0]
    rer=rer.drop('Unnamed: 0', axis=1)
    er=er.append(rer,ignore_index=True)   
stock=er.sort_values(by='% Изм. цены')
stock=stock.dropna(subset=['% Изм. цены'])
a=[]
for i in range(425):
    a.append(stock.values[i][0].partition('Потенциал')[0])
stock['Инструмент']=a
stock
```
